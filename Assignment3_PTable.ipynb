Define properties of each element. The output of this function gives all information about each element.

```python
def Properties(element_dictionary):
    print("Name: " + element_dictionary[0])
    print("Atomic Number: " + str(element_dictionary[1]))
    print("Row: " + str(element_dictionary[2]))
    print("Column: " + str(element_dictionary[3]))
    print("\n")
```
