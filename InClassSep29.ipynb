```python
movie_name = 'The Godfather'
year = '1972'
genre = 'Crime, Drama'
actor = 'Maelon Brando, Al Pacino, Robert Duvall'
review = ''
flag = 'False'
borrower = ''
counter = 0
actor_name1 = 'Maelon Brando'
date_of_birth1 = '1980'
nationality1 = 'USA'
actor_name2 = 'Al Pacino'
date_of_birth2 = '1981'
nationality2 = 'USA'
actor_name3 = 'Robert Duvall'
date_of_birth3 = '1982'
nationality3 = 'USA'
```

```python
class Movie():

    def __init__(self, movie_name, year, genre, actor, review, flag, borrower, counter):
        self.movie_name = movie_name
        self.year = year
        self.genre = genre
        self.actor = actor
        self.review = review
        self.flag = flag
        self.borrower = borrower
        self.counter = counter

    def watching_the_movie(self):
        self.counter  = self.counter + 1

    def writing_a_review(self, review):
        self.review = review

    def change_genres(self, genres):
        self.genres = genres

    def change_year(self, year):
        self.year = year
    
    def change_actor(self, actor):
        self.actor = actor

    def borrowing_the_movie(self, borrower):
        self.borrowed = True
        self.borrower = borrower

    def returning_the_movie(self):
        self.borrowed = False
        self.borrower = ''

    def Whatching_Movie(self):
        print("Movie Name:",self.movie_name)
        print("Year:",self.year)
        print("Genre:",self.genre)
        print("List of Actors:",self.actor)
        print("Review:", self.review)
        print("Flag:",self.flag)
        print("Borrower:",self.borrower)
        print("Counter:",self.counter)
```

```python
class Actor:

    def __init__(self, actor_name, date_of_birth, nationality):
        self.actor_name = actor_name
        self.date_of_birth = date_of_birth
        self.nationality = nationality

    def set_actor_name(self,actor_name):
        self.actor_name = actor_name

    def set_date_of_birth(self,date_of_birth):
        self.date_of_birth = date_of_birth

    def set_nationality(self,nationality):
        self.nationality = nationality

    def ShowActor(self):
        print("Actor Name:",self.actor_name,", Date of Birth:",self.date_of_birth,", Nationality:",self.nationality)
        
```

```python
if __name__ == "__main__":

    while True:
        print("\n ** Main Menu ** \n")
        print("0. movie information: ")
        print("1. watching movie: ")
        print("2. actor information: ")
        print("3. Write the Movie Review: ")
        print("4. Borrow Movie:")
        print("5. Return Movie:")
        print("6. Exit")
        option = int(input('Enter your Option: '))
        if (option == 0):
            movie = Movie(movie_name, year, genre, actor, review, flag, borrower,counter)
            movie.Whatching_Movie()
        
        elif (option == 1):
            counter += 1
            movie = Movie(movie_name, year, genre, actor, review, flag, borrower,counter)
            movie.Whatching_Movie()

        elif (option == 2):
            actor1 = Actor(actor_name1, date_of_birth1, nationality1)
            actor1.ShowActor()
            actor2 = Actor(actor_name2, date_of_birth2, nationality2)
            actor2.ShowActor()
            actor3 = Actor(actor_name3, date_of_birth3, nationality3)
            actor3.ShowActor()

        elif (option == 3):
            review = input("Please enter the review:")
            movie = Movie(movie_name, year, genre, actor, review, flag, borrower, counter)
            movie.Whatching_Movie()

        elif (option == 4):
            if flag == 'False':
                flag = 'True'
                borrower = input("Enter the borrower name:")
                movie = Movie(movie_name, year, genre, actor, review, flag, borrower, counter)
                movie.Whatching_Movie()
            else:
                print("Already borrowed by",borrower)

        elif (option == 5):
            if flag == 'True':
                flag = 'False'
                #borrower = ''
                movie = Movie(movie_name, year, genre, actor, review, flag, borrower, counter)
                movie.Whatching_Movie()
                print(borrower, "returned movie")
                borrower = ''
            else:
                print("Not Borrowed")
            
        elif (option == 6):
            print("Exiting")
            break
            
        else:
              print("Wrong entry, Please re-enter\n")
```
